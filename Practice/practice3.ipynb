```python
import math
import random
```

```python
cities={"A":(1,2),"B":(3,4),"C":(2,4),"D":(4,7),"E":(5,3),"F":(3,5),"G":(2,6),"H":(4,1)}
populationSize=100
generation=500
mutationRate=0.02
```

```python
def distance(route):
    total=0
    for i in range(len(route)):
        x1,y1=cities[route[i]]
        x2,y2=cities[route[(i+1)%len(route)]]
        dist=math.sqrt((x2-x1)**2 + (y2-y1)**2)
        total+=dist
    return total
```

```python
population=[]
for _ in range(populationSize):
    population.append(random.sample(list(cities.keys()),len(cities)))
```

```python
bestRoute=None
shortestDistance=float("inf")
```

```python
for gen in range(1,generation+1):

    selected=[]

    for _ in range(populationSize):
        r1=random.choice(population)
        r2=random.choice(population)
        if (distance(r1)<distance(r2)):
            selected.append(r1)
        else:
            selected.append(r2)

    population=[]

    for child in selected:
        mutated=child[:]
        for i in range(len(mutated)):
            if (random.random()>mutationRate):
                j=random.randint(0,len(child)-1)
                mutated[i],mutated[j]=mutated[j],mutated[i]
        population.append(mutated)

    current=min(population,key=distance)
    if(distance(current)<shortestDistance):
        shortestDistance=distance(current)
        bestRoute=current

    if (gen%100==0):
        print(f"Generation : {gen} | Best Distance : {shortestDistance:.4f} | Best Route : {bestRoute}\n")

print(f"The best route is {bestRoute} and the minimum distance is {shortestDistance:.4f}\n")
```
