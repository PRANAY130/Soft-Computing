# Fuzzy Set Operations

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Union using Bell shaped membership function following fuzzy sets


```python
def bellShaped(x,a,b,c):
    return (1/(1+(np.abs((x-c)/a)**(2*b))))

x=np.linspace(0,10,1000)

setA=bellShaped(x,1,2,4)
setB=bellShaped(x,2,4,6)

union=np.maximum(setA,setB)

plt.plot(x,setA,linestyle="dashed",color="red")
plt.plot(x,setB,linestyle="dashed",color="green")
plt.plot(x,union,color="blue")
plt.fill_between(x,union,color="yellow",alpha=0.4)
plt.grid()
plt.show()
```

### Max Min Composition

```python
# setAB=np.array([])
# setBC=np.array([])

setAB = np.array([[0.5, 0.7, 0.2],
                  [0.8, 0.6, 0.9]])

setBC = np.array([[0.6, 0.3],
                 [0.9, 0.4],
                 [0.5, 0.8]])


result=[]
for i in range (len(setAB)):
    el=[]
    for j in range(len(setBC[0])):
        re=[]
        for k in range (len(setBC)):
            e=np.minimum(setAB[i][k],setBC[k][j])
            re.append(e)
        el.append(max(re))
    result.append(el)

print(np.array(result))
            
```

### Algebric Difference using Gaussian membership function following sets

```python
def gaussian(x,a,c):
    return (np.exp(-(0.5)*((x-a)**2)/(c**2)))

x=np.linspace(0,10,1000)

setA=gaussian(x,4,1)
setB=gaussian(x,5,0.78)

algebricDif=np.maximum(0,setA-setB)

plt.plot(x,setA,linestyle="dashed",color="blue")
plt.plot(x,setB,linestyle="dashed",color="red")
plt.plot(x,algebricDif,color="yellow")
plt.fill_between(x,algebricDif,color="yellow",alpha=0.5)
plt.grid()
plt.show()
```
