# Young and Old fuzzy Problem

```python
def bellmembership(x,a,b,c):
    return 1 / (1 + abs((( x - c ) / a )) ** ( 2 * b ))

x=int(input("Enter the age to be checked:"))

young=bellmembership(x,20,2,0)
old=bellmembership(x,30,3,100)

moreOrLessYoung=young**0.5
youngButnotTooyoung=min(young,1-young**2)
notYoungandnotOld=min(1-young,1-old)
extremelyOld=old**2

print("More or less young:",moreOrLessYoung)
print("Young but not too young:",youngButnotTooyoung)
print("Not young and not old:",notYoungandnotOld)
print("Extremely Old:",extremelyOld)
```

# Humidity and Temperature

```python
temp={}
humid={}

print("Take input of the temperatures")
n=int(input("Enter the number of entries of temperature:"))

for i in range(n):
    temperature=float(input("Enter the temperature value:"))
    mem=float(input("Enter the membership value"))
    temp[temperature]=mem

print("Take input of the humidity")
n=int(input("Enter the number of entries of humidity:"))

for i in range(n):
    humidity=float(input("Enter the humidity value:"))
    mem1=float(input("Enter the membership value"))
    humid[humidity]=mem1

thresTemp=float(input("Enter the threshold temperature"))
threshumid=float(input("Enter the threshold humidity"))

for t in temp:
    for h in humid:
        if(humid[h]>=threshumid and temp[t]>=thresTemp):
            print(f"Temperature,Humidity: : {t},{h} Membership: {max(temp[t],humid[h])} ")
```
