```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def trapezoidalMembership(x,a,b,c,d):
    return np.maximum(0,np.minimum(np.minimum(1,(x-a)/(b-a)),((d-x)/(d-c))))
```

```python
x=np.linspace(0,10,1000)

y=trapezoidalMembership(x,2,4,6,7)

plt.plot(x,y)
```

# Young and Old fuzzy Problem

```python
def bellmembership(x,a,b,c):
    return 1 / (1 + abs((( x - c ) / a )) ** ( 2 * b ))

x=int(input("Enter the age to be checked:"))

young=bellmembership(x,20,2,0)
old=bellmembership(x,30,3,100)

moreOrLessYoung=young**0.5
youngButnotTooyoung=min(young,1-young**2)
notYoungandnotOld=min(1-young,1-old)
extremelyOld=old**2

print("More or less young:",moreOrLessYoung)
print("Young but not too young:",youngButnotTooyoung)
print("Not young and not old:",notYoungandnotOld)
print("Extremely Old:",extremelyOld)
```
